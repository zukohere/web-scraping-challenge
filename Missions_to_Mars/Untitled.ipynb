```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
import pandas as pd
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
### NASA Mars News
url = "https://redplanetscience.com/"
browser.visit(url)
```

```python
headline_dict={}
teaser_dict = {}
for page in range(0, 10):
    page = page+1
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    headlines = soup.find_all('div', class_='content_title')
    teasers = soup.find_all('div', class_='article_teaser_body')
    h_count = 1
    
    for headline in headlines:
        for teaser in teasers: 
            print(f"Page: {page}")
            print(f"headline: {h_count}")
            print(headline.text)
            print(teaser.text)
            print("-----------")
            headline_dict[(page,h_count)]=headline.text
            teaser_dict[(page,h_count)]=teaser.text
            h_count = h_count+1
```

```python
news_title = headline_dict[(1,1)]
news_p = teaser_dict[(1,1)]
print(news_title)
print(news_p)
```

```python
### Mars Facts
facts_url = "https://galaxyfacts-mars.com"
tables = pd.read_html(facts_url)
tables
```

```python
# Merge the two list elements
df = tables[0].merge(tables[1],how="outer")
df
```

```python
# drop columns related to Earth because that planet is so over. Drop row 0.
df = df[[0,1]]
mars_df = df.drop([0])
df
```

```python
# rename the columns, reset the index.
mars_df = mars_df.rename(columns={0: "Parameter", 1:"Mars Value"})
mars_df = mars_df.reset_index(drop=True)
mars_df
```

```python
# Create html string.
html_table = mars_df.to_html()
html_table = html_table.replace('\n', '')
html_table
```

```python
# save to file
mars_df.to_html('mars_facts_table.html')
```
